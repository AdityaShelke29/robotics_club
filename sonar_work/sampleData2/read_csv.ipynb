```python
# Classic Imports 
import pandas as pd
import numpy as np
from pathlib import Path

dataFile = "./hydro4_3.5mA_close.csv"

# The load_hydrophone_data function opens a CSV file. 
def load_hydrophone_data():
    return pd.read_csv(Path(dataFile), on_bad_lines='skip', skiprows = 9)

hydrophoneData = load_hydrophone_data()
hydrophoneData.head()                                  # Gives first couple of lines of the csv
```

```python
sampleHydrophoneData = hydrophoneData.iloc[:, :]
sampleHydrophoneData = pd.DataFrame(sampleHydrophoneData)

sampleHydrophoneData.plot(x = 0)
```

```python
sampleHydrophoneData = hydrophoneData.iloc[3700:4000, :]
sampleHydrophoneData = pd.DataFrame(sampleHydrophoneData)

sampleHydrophoneData.plot(x = 0)
```
